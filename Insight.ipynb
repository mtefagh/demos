```julia
using Plots
T = 1000;
S = 0.1;
user = 10000;
agent = 100;
M = 100_000;
userTX = M/user;
agentTX = M/agent;
fee = 100_000;
payU = M*fee/2;
payA = M*fee/2;
transA = 0;
plotly();
for t in 1:2*T
    volU = sum([rand(0:userTX) for i in 1:user]);
    if t > T
        transA = transA + sum([rand(0:agentTX) for i in 1:agent]);
        if t % 10 == 0
            volA = copy(transA);
            transA = 0;
        else
            volA = 0;
        end
    else
        volA = sum([rand(0:agentTX) for i in 1:agent]);
    end
    payU = [payU; volU*fee[end]];
    payA = [payA; volA*fee[end]];
    fee = [fee; fee[end]*(1+S*((volU + volA)/M - 1))];
end

plot(fee, label = "Transaction Fee Pricing");
xlabel!("Block")
```

```julia
plot(payA, label = "Total Agents Payment");
plot!(payU, label = "Total Users Payment");
xlabel!("Block")
```

```julia
println("Total Agents Payment: ", sum(payA))
println("Total Users Payment: ", sum(payU))
```
