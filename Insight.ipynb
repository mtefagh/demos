```julia
using Plots
T = 1000;
S = 0.125;
user = 10000;
agent = 100;
M = 100_000;
userTX = M/user;
agentTX = M/agent;
fee = 100_000;
payU = M*fee/2;
payA = M*fee/2;
plotly();
```

```julia
transA = 0;
for t in 1:2*T
    volU = sum([rand(0:userTX) for i in 1:user]);
    if t > T
        transA = transA + sum([rand(0:agentTX) for i in 1:agent]);
        if t % 10 == 0
            volA = copy(transA);
            transA = 0;
        else
            volA = 0;
        end
    else
        volA = sum([rand(0:agentTX) for i in 1:agent]);
    end
    payU = [payU; volU*fee[end]];
    payA = [payA; volA*fee[end]];
    fee = [fee; fee[end]*(1+S*((volU + volA)/M - 1))];
end

plot(fee, linewidth = 2, title = "Transaction Fee Pricing");
xlabel!("Block")
```

```julia
plot(payU, linewidth = 2, title = "Total Users Payment");
xlabel!("Block")
```

```julia
plot(payA, linewidth = 2, title = "Total Agents Payment");
xlabel!("Block")
```
